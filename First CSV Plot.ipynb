```python
import os
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

#os.environ['PATH'] = os.environ['PATH'] + ";C:\\Users\\q533604\\AppData\\Local\\Programs\\MiKTeX\\miktex\\bin" 
label_size = 8
# plt.style.use('fivethirtyeight')
plt.rc("text", usetex=True)
plt.rc("text.latex", preamble=r'\usepackage{amsmath}')
 
pr ={"axes.labelsize": 8,               # LaTeX default is 10pt font.
    "font.size": 8,
    "legend.fontsize": 8,               # Make the legend/label fonts
    "xtick.labelsize": 8,               # a little smaller
    "ytick.labelsize": 8,
    'figure.figsize': (3.35, 9 * 3.35 / 16),
    "errorbar.capsize": 2.5,
    "font.family": "serif",
    "font.serif": [],                   # blank entries should cause plots
    "font.sans-serif": [],    
    }
for k, v in pr.items():
    mpl.rcParams[k] = v
   
mpl.rcParams["font.family"] = "computer modern roman"
# mpl.rcParams["font.serif"] = ["STIX"]
mpl.rcParams["mathtext.fontset"] = "stix"
```

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define the piecewise approximation function
def piecewise_approx(x):
    y = np.zeros_like(x)
    y[x > 0.5] = 1
    y[(x > -0.5) & (x <= 0.5)] = x[(x > -0.5) & (x <= 0.5)] + 0.5
    return y

# Create an array of x values
x = np.linspace(-2.5, 2.5, 1000)

# Generate y values for both functions
y_sigmoid = sigmoid(x)
y_approx = piecewise_approx(x)

# Plot both functions
plt.figure(figsize=(8, 4))
plt.plot(x, y_sigmoid, label='Sigmoid Function')
plt.plot(x, y_approx, label='Piecewise Linear Approximation', linestyle='--')

plt.grid(True, which='both', axis='x', linestyle='--', linewidth=1)
# Add axis
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
# Add x and y labels
plt.xticks([-0.5, 0.5], ['-0.5', '0.5'])
plt.yticks([0, 1], ['0', '1'])

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.savefig('sigmoid.png')
# Show the plot
plt.show()
```
